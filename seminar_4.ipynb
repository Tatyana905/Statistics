# Задача 4 

Рост взрослого населения города X имеет **нормальное распределение**.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост?

**Дано:** 

μ(среднее) = 174 см, 

σ(среднее квадратичное отклонение) = 8 см, 

**Решение:** 

Наше имеющиеся распределение нужно поместить в масштабы нормального стандартного распределения по формуле:
$$ Z = \frac{(X - μ)} {σ}$$

Далее пользуюясь таблицей накопленного нормального распределения получим значения для каждого значения.

```python
sigma = 8
mu = 174
def Z(x, sigma, mu):
  return (x-mu)/ sigma

values = [182, 190, 166, 190, 166, 182, 158, 190, 150, 190, 150, 198, 166]

for i in values:
  z = Z(i, sigma, mu)
  print(f"Для X = {i}, Z = {z}")
```

а) больше 182 см, Z = 1

**Ответ: P(> 182) = 1 - 0,84134 = 0,15866**

б) больше 190 см, Z = 2.0

**Ответ: P(> 190) = 1 - 0.97725 = 0,02275**

в) от 166 см до 190 см, Z(166) = -1.0, Z(190) = 2.0

**Ответ: P(166 < X < 190) = 0.97725 - 0.1587 = 0,81855**

г) от 166 см до 182 см, Z(166) = -1.0, Z(182) = 1.0

**Ответ: P(166 < X < 182) = 0,84134 - 0.1587 = 0,68264**

д) от 158 см до 190 см, Z(158) = -2.0, Z(190) = 2.0

**Ответ: P(158 < X < 190) = 0.97725 - 0.0228 = 0,95445**

е) не выше 150 см или не ниже 190 см, Z(150) = -3.0, Z(190) = 2.0

**Ответ: P(<150 ИЛИ >190) = 0.0014 + (1 - 0.97725) = 0,02415**

ё) не выше 150 см или не ниже 198 см, Z(150) = -3.0, Z(198) = 3.0

**Ответ: P(<150 ИЛИ >198) = 0.0014 + (1 - 0.99865) = 0,00275**

ж) ниже 166 см, Z(166) = - 1.0

**Ответ: P(<166) = 0,1587**
