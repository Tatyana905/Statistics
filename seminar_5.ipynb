# –ó–∞–¥–∞–Ω–∏–µ 2

–ü—Ä–æ–≤–µ–¥–∏—Ç–µ —Ç–µ—Å—Ç –≥–∏–ø–æ—Ç–µ–∑—ã. 

*–£—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç—Å—è, —á—Ç–æ —à–∞—Ä–∏–∫–∏ –¥–ª—è –ø–æ–¥—à–∏–ø–Ω–∏–∫–æ–≤, –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã–µ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–º —Å—Ç–∞–Ω–∫–æ–º, –∏–º–µ—é—Ç —Å—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä 17 –º–º.*

–ò—Å–ø–æ–ª—å–∑—É—è –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π —Å **Œ±=0,05**, –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —ç—Ç—É –≥–∏–ø–æ—Ç–µ–∑—É, –µ—Å–ª–∏ –≤ –≤—ã–±–æ—Ä–∫–µ –∏–∑ **n=100** —à–∞—Ä–∏–∫–æ–≤ **—Å—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä –æ–∫–∞–∑–∞–ª—Å—è —Ä–∞–≤–Ω—ã–º 17.5 –º–º**, –∞ **–¥–∏—Å–ø–µ—Ä—Å–∏—è –∏–∑–≤–µ—Å—Ç–Ω–∞ –∏ —Ä–∞–≤–Ω–∞ 4 –∫–≤. –º–º.**

–î–∞–Ω–æ:
  $$n = 100,  \overline{x} = 17,5 –º–º, Œº (—Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ) =  17  –º–º$$
  $$D = 4 –∫–≤. –º–º., Œ± (—É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç. –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏) = 0,05$$

–¢–∞–∫ –∫–∞–∫ –Ω–∞–º –¥–∞–Ω–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, –¥–ª—è —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏—è –≥–∏–ø–æ—Ç–µ–∑—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º Z ‚Äì–∫—Ä–∏—Ç–µ—Ä–∏–π:

  $$Z_h = \frac{\overline{x} - Œº} {\frac{\sqrt{ùûº}}{\sqrt{{n}}}}$$
  $$Z_h = \frac{17,5 - 17} {\frac{\sqrt{4}}{\sqrt{100} }}$$

```python
Z = (17.5 - 17)/ (4**0.5/100**0.5)
print(f'Z —Ä–∞—Å—á–µ—Ç–Ω—ã–π —Ä–∞–≤–µ–Ω {Z}.')
```

Z_—Ç–∞–±–ª–∏—á–Ω—ã–π Œ± = 0.05 = -1,645

Z_—Ç–∞–±–ª–∏—á–Ω—ã–π Œ± = 0.95 = 1,645

$$Z_—Ä–∞—Å—á–µ—Ç–Ω—ã–π > Z_—Ç–∞–±–ª–∏—á–Ω—ã–π$$

**–û—Ç–≤–µ—Ç: –û—Ç–≤–µ—Ä–≥–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É, —á—Ç–æ —à–∞—Ä–∏–∫–∏ –¥–ª—è –ø–æ–¥—à–∏–ø–Ω–∏–∫–æ–≤, –∏–∑–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã–µ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–º —Å—Ç–∞–Ω–∫–æ–º, –∏–º–µ—é—Ç —Å—Ä–µ–¥–Ω–∏–π –¥–∏–∞–º–µ—Ç—Ä 17 –º–º.**

# –ó–∞–¥–∞–Ω–∏–µ 3
–ü—Ä–æ–≤–µ–¥–∏—Ç–µ —Ç–µ—Å—Ç –≥–∏–ø–æ—Ç–µ–∑—ã. 

*–ü—Ä–æ–¥–∞–≤–µ—Ü —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–π –≤–µ—Å –ø–∞—á–∫–∏ –ø–µ—á–µ–Ω—å—è —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 200 –≥.*


–ò–∑ –ø–∞—Ä—Ç–∏–∏ –∏–∑–≤–ª–µ—á–µ–Ω–∞ –≤—ã–±–æ—Ä–∫–∞ –∏–∑ 10 –ø–∞—á–µ–∫. 

–í–µ—Å –∫–∞–∂–¥–æ–π –ø–∞—á–∫–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
–ò–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –∏—Ö –≤–µ—Å–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ. –í–µ—Ä–Ω–æ –ª–∏ —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–∞, –µ—Å–ª–∏ —É—á–∏—Ç—ã–≤–∞—Ç—å, —á—Ç–æ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 99%? (–ü—Ä–æ–≤–µ—Å—Ç–∏ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ç–µ—Å—Ç).

**1. –°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É:** 

  $$ H_0: Œº_1 = Œº_0 = 200 –≥. $$

  $$ H_1: Œº_1 \neq Œº _0$$

**2. –í—ã–±–æ—Ä —É—Ä–æ–≤–Ω—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ Œ±(–∞–ª—å—Ñ–∞):**

Œ±(–∞–ª—å—Ñ–∞) = 0,01(1%) - –ø–æ —É—Å–ª–æ–≤–∏—é

**3. –í—ã–±–æ—Ä —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –∫—Ä–∏—Ç–µ—Ä–∏—è t - –∫—Ä–∏—Ç–µ—Ä–∏–π (–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞), —Ç.–∫. ùûº –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞**

**4. –†–∞—Å—á–µ—Ç –Ω–∞–±–ª—é–¥–∞–µ–º–æ–≥–æ –∫—Ä–∏—Ç–µ—Ä–∏—è –ø–æ –≤—ã–±–æ—Ä–∫–µ**

```python
import numpy as np
```

```python
x =np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190]) 
```

```python
d1 = np.var(x, ddof=1)
d1
```

```python
m1 = np.mean(x)
m1
```

D(X) = 19.834, x(—Å—Ä–µ–¥. –∞—Ä–∏—Ñ–º.) = 198.5, Œº = 200

–ü—Ä–æ–∏–∑–≤–æ–¥–∏–º —Ä–∞—Å—á–µ—Ç t - –∫—Ä–∏—Ç–µ—Ä–∏–π (–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞) –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

$$t = \frac{\overline{x} - Œº} {\sqrt{\frac{D(X)}{n}}}$$


```python
Œº = 200
t = (m1 - Œº) /np.sqrt(d1/len(x)) 
t
```

**5. –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—á–Ω–æ–≥–æ –∏ –Ω–∞–±–ª—é–¥–∞–µ–º–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è**

–ò—â–µ–º t_—Ç–∞–±–ª–∏—á–Ω–æ–µ. 

–¢–∞–∫ –∫–∞–∫ –ø–æ —É—Å–ª–æ–≤–∏—é —Å–∫–∞–∑–∞–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ç–µ—Å—Ç - –∏—â–µ–º 1% –≤ —Ç–∞–±–ª–∏—Ü–µ –¥–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–µ–≥–æ –∫—Ä–∏—Ç–µ—Ä–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞  

t(Œ± = 0,01, —Å—Ç–µ–ø–µ–Ω–∏ —Å–≤–æ–±–æ–¥—ã 9) = 3,25

**6. –í—ã–≤–æ–¥**

t_—Ä–∞—Å—á–µ—Ç–Ω–æ–µ = -1,065

t_—Ç–∞–±–ª–∏—á–Ω–æ–µ = 3,25

–ü—Ä–∏ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–º —Ç–µ—Å—Ç–µ –∏—Å–ø–æ–ª—å–∑—É–µ–º –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è

t_—Ç–∞–±–ª–∏—á–Ω–æ–µ > t_—Ä–∞—Å—á–µ—Ç–Ω–æ–µ

**–û—Ç–≤–µ—Ç: –ø—Ä–∏–Ω–∏–º–∞–µ–º –Ω—É–ª–µ–≤—É—é —Ç–µ–æ—Ä–∏—é, –Ω–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –≤–∞–∂–Ω—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π.**

# –ó–∞–¥–∞—á–∞ 4
–ï—Å—Ç—å –ª–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –≤ —Ä–æ—Å—Ç–µ –¥–æ—á–µ—Ä–µ–π –∏ –º–∞—Ç–µ—Ä–µ–π?

**1. –°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É:** 

–ù–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –≤–∞–∂–Ω—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π:

Œ±(–∞–ª—å—Ñ–∞) = 0,05(5%) 

**3. –í—ã–±–æ—Ä —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –∫—Ä–∏—Ç–µ—Ä–∏—è t - –∫—Ä–∏—Ç–µ—Ä–∏–π (–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞), —Ç.–∫. ùûº –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞**

```python
import scipy.stats as stats
```

```python
mothers = np.array([172, 177, 158, 170, 178,175, 164, 160, 169]) 
daughters = np.array([173, 175, 162, 174, 175, 168, 155, 170, 160])
```

```python
stats.ttest_rel(mothers,daughters)
```

* pvalue = 0,559 > Œ±(–∞–ª—å—Ñ–∞) = 0,05
